Forming the clusters using Elbow Method

```python
#Forming the clusters
from sklearn.cluster import KMeans
kmeans_model=KMeans(n_clusters=5)
y=kmeans_model.fit_predict(df[['age','Income($)']])
y
```

```python
#Calculating the sse values from 1 to 10 clusters
sse=[]
for i in range(1,11):
    kmeans_model=KMeans(n_clusters=i,init='k-means++',random_state=20)
    kmeans_model.fit(df[['age','Income($)']])
    sse.append(kmeans_model.inertia_)
```

```python
#Plotting the sse values
plt.plot(range(1,11),sse)
plt.title('Elbow Method')
plt.xlabel('Number of Clusters')
plt.ylabel('SSE')
plt.show()
```

```python
From the above grapph it is observed that 3 clusters are enough for clusters
```
